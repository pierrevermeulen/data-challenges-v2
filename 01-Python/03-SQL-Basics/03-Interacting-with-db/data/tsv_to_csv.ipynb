# IMDB movie database creation

datasets:

- https://datasets.imdbws.com/name.basics.tsv.gz
- https://datasets.imdbws.com/title.basics.tsv.gz
- https://datasets.imdbws.com/title.crew.tsv.gz
- https://datasets.imdbws.com/title.ratings.tsv.gz

```python
import pandas as pd
```

```python
movie_df = pd.read_csv('./title.ratings.tsv', sep='\t')
movie_df.head()
```

```python
movie_df = movie_df[movie_df["numVotes"] > 10000]
movie_df.head()
```

```python
tmp_movie_df = pd.read_csv('./title.crew.tsv', sep='\t')
tmp_movie_df.head()
```

```python
movie_df.set_index("tconst")
tmp_movie_df.set_index("tconst")
```

```python
movie_df = movie_df.merge(tmp_movie_df, how="inner")
```

```python
movie_df.head()
```

```python
tmp_movie_df = pd.read_csv('./title.basics.tsv', sep='\t')
movie_df.set_index("tconst")
tmp_movie_df.set_index("tconst")
movie_df = movie_df.merge(tmp_movie_df, how="inner")
movie_df
```

```python
movie_df["directors"] = movie_df["directors"].apply(lambda x: x.split(",")[0])
movie_df["directors"]
```

```python
movie_df = movie_df.drop(columns=["writers", "titleType", "isAdult", "originalTitle", "endYear"])
```

```python
movie_df
```

```python
tmp_movie_df = pd.read_csv('./name.basics.tsv', sep='\t')
movie_df = movie_df.merge(tmp_movie_df, how="inner", left_on="directors", right_on="nconst")
movie_df
```

```python
movie_df = movie_df.drop(columns=["primaryProfession", "knownForTitles"])
movie_df
```

```python
movie_df.to_csv("movies.csv")
```
